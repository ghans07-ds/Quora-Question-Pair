```python
import pandas as pd
import pickle
```

```python
df=pd.read_csv("E:\DATA Science\Datasets\Kaggle\Quora//train_4.csv")
```

```python
df.columns
```

```python
features_col=['len_def',
       'total_words','common_token','word_share','stop_count',
       'common_word', 'cwc_min', 'cwc_max', 'csc_min', 'csc_max', 'ctc_min',
       'ctc_max', 'last_word_eq', 'first_word_eq', 'fuzz_ratio',
       'partial_ratio', 'token_sort_ratio']
target_col=['is_duplicate']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df[features_col].values
y=df[target_col]from sklearn.model_selection import train_test_split.values

X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.30,random_state=10)
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest_model=RandomForestClassifier(random_state=10)

random_forest_model.fit(X_train,Y_train.ravel())
```

```python
pickle.dump(random_forest_model,open('E:\DATA Science\Datasets\Kaggle\Quora/dqi.pickle','wb'))
```

```python
predict_train_data=random_forest_model.predict(X_test)

from sklearn import metrics

print("Accuracy={0:.3f}".format(metrics.accuracy_score(Y_test,predict_train_data)))
```
